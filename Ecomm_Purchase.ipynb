```python
#Q15. What are the top 5 most popular email providers (e.g. gmail.com, yahoo.com, etc...)
#Method 01: looping

l1 = []
for emails in df['Email']:
    l1.append(emails.split('@')[1])

df['Email_provider'] = l1
print('Top Email providers are:')
df['Email_provider'].value_counts().head(5)
```

```python
#Q15. What are the top 5 most popular email providers (e.g. gmail.com, yahoo.com, etc...)
#Method 02: Using lambda function
print('Top Email providers are:')
df['Email'].apply(lambda x:x.split('@')[1]).value_counts().head(5)
```
